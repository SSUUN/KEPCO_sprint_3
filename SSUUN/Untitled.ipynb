```python
import pandas as pd
import os
t=[]
for i in os.listdir("data/"):
    t.append(pd.read_csv(f"data/{i}",encoding="cp949"))
y=pd.merge(t[0],t[1],on="행정구역",how="outer")
#y=pd.merge(y,t[2],on="행정구역",how="outer")
y=y[y["행정구역"].apply(lambda x: "출장소" not in x)].dropna()
y
#y[y["행정구역"].apply(lambda x:"출장소" not in x)]
    #pd.concat(t,axis=1)
```

```python
# 도, 시 이름 2글자로
def f(x):
    x=x.split()[0]
    if "도" in x:
        if len(x)==3:
            return x[:2]
        elif len(x)==4:
            return x[0]+x[2]
        else:
            return "제주"
    else:
        return x[:2]
    
y["metro"]=y["행정구역"].apply(f)

# 시군구 뽑기
def f1(x):
    s=x.split()
    if len(s)>2:
        return s[1]
    else:
        if "세종" in x:
             return x
y["city"]=y["행정구역"].apply(f1)
```
